### Ex 1. Vector recycling
It is the concept of aligning vectors and perform a sequence of individual operations. It will repeat the shorter vectors until it is as the same length as the longer vector. However the length of the short has to be multiple of the long.

```r
r <- 1:8
x <- 1:2
c <- r + x
print(c)
```

```r
x <- 1:3
c <- r + x
print(c)
```

### Ex 2. Inner multiplication
It multiplies and sum up the result into a value on two identical vectors or vectors with the same length.

```r
a <- 1:6
b <- 2:7
d <- a %*% b
print(d)
```

```r
d <- a %*% a
print(d)
```

### Ex 3. Outer multiplication
It is similiar to matrix multiplication or the product operation.

```r
d <- a %o% a
print(d)
```

### 4. Functions
* sample() to pick random range of number from a list or vector
* set.seed() to set a constant as a key to generate random sequence
* round() to round up number
* factorial() to calculate the factorial of a vector
* mean() is to find the average by sum all elements and divide by the number elements
* ls() is to list all the registered and initialized variables

```r
# set.seed(1)
sample(1:6, size=2)
# set.seed(1)
sample(1:6, size=2)
seq(1:5)
rep(4:6, 3)
round(4.643)
round(95.23662, 3)
factorial(5)
mean(2:8)
ls()
```

### 6. Write a program to calculate the BMI rate?Get the user input & Result should be in integer. 

```r
weight <- readline(prompt = "Enter the weight: ")
weight <- as.integer(weight)
height <- readline(prompt = "Enter the height: ")
height <- as.integer(height)
bmi = weight / ((height/100)^2)
print(round(bmi, 2))
```

### 7. Create a function to calculate the BMI Rating? Result should be in integer?

```r
BMI.function = function(w, h) {
    result = w / ((h/100)^2)
    print(round(result, 2))
}

weight <- readline(prompt = "Enter the weight: ")
weight <- as.integer(weight)
height <- readline(prompt = "Enter the height: ")
height <- as.integer(height)
BMI.function(weight, height)
```
